# **Disease Prediction Model**

#### The objective of this project is to develop a predictive model that can accurately classify individuals into specific diseases categories based on their health attributes

#### **Importing modules**

```python
import pandas as pd
```

#### **Importing csv files**

```python
train_data = pd.read_csv("C:/Users/User/Downloads/Train_data.csv")
test_data = pd.read_csv("C:/Users/User/Downloads/test_data.csv")
```

#### **Merging the two data into one**

```python
combined_data = pd.concat([train_data, test_data], axis=0)
combined_data.reset_index(drop=True, inplace=True)
```

## **Model**

#### **Importing modules**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

#### **Dropping the Disease column and assigning to X**

```python
X = combined_data.drop(['Disease'], axis=1)
```

#### **Extracting Target Variable**

```python
Y = combined_data['Disease']
```

#### **Spliting Data into training and testing set**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

#### **Initializing Random Forest Classifier**

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

#### **Training the Model**

```python
model.fit(X_train, y_train)
```

#### **Predicting the test set**

```python
y_pred = model.predict(X_test)
y_pred
```

#### **Accuracy**

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

#### **Precision Score**

```python
precision = precision_score(y_test, y_pred, average= 'macro', zero_division=0)
precision
```

#### **Recall Score**

```python
recall = recall_score(y_test, y_pred, average= 'macro', zero_division=0)
recall
```

#### **F1 Score**

```python
f1 = f1_score(y_test, y_pred, average= 'macro', zero_division=0)
f1
```

#### **Confusion Matrix**

```python
cm = confusion_matrix(y_test, y_pred)
cm
```
